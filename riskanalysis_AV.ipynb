**F-N Curve**

Data Set: https://www.kaggle.com/datasets/ibriiee/tesla-autonomous-deaths-data-updated-2023#

According to Data:
Number of Fatalities per Accident:

*   1 fatality: 248 accidents
*   2 fatalities: 40 accidents
*   3 fatalities: 7 accidents
*   4 fatalities: 3 accidents

Cumulative Frequency:

*   1 fatality: 248
*   2 fatalities: 288 (248 + 40)
*   3 fatalities: 295 (288 + 7)
*   4 fatalities: 298 (295 + 3)

FN Curve Plot:
*   X-axis: Number of Fatalities per Accident - [1, 2, 3, 4]
*   Y-axis: Cumulative Frequency of Accidents - [248, 288, 295, 298]











```python
import matplotlib.pyplot as plt

# Provided data
number_of_fatalities = [1, 2, 3, 4]
frequency = [248, 40, 7, 3]

# Calculate the cumulative frequency
cumulative_frequency = [sum(frequency[:i+1]) for i in range(len(frequency))]

# Plot the FN curve
plt.figure(figsize=(10, 6))
plt.plot(number_of_fatalities, cumulative_frequency, marker='o', linestyle='-')
plt.xlabel('Number of Fatalities per Accident')
plt.ylabel('Cumulative Frequency of Accidents')
plt.title('FN Curve for Tesla Autonomous Vehicle Fatalities')
plt.grid(True)
plt.show()
```
